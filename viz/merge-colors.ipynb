## Merging Team Colors from ESPN/`ncaahoopR` with Kaggle Names for Viz Purposes

```python
import os
import pandas as pd
```

### Women's

```python
os.chdir("../data/women/")
teams = pd.read_csv("WTeams.csv")
team_spellings = pd.read_csv("WTeamSpellings.csv")
w_team_colors = pd.read_csv("ncaa_colors.csv")
```

```python
teams_merge = teams.merge(w_team_colors[['kaggle_name', 'primary_color', 'secondary_color']],
                          how = 'left', left_on = 'TeamName', right_on = 'kaggle_name')
```

### Women's teams without colors

```python
teams_merge[pd.isna(teams_merge['primary_color'])][['TeamID', 'TeamName']]
```

## Men's

```python
os.chdir("../men/")
teams = pd.read_csv("MTeams.csv")
team_spellings = pd.read_csv("MTeamSpellings.csv")
m_team_colors = pd.read_csv("ncaa_colors.csv")
```

```python
m_teams_merge = teams.merge(m_team_colors[['kaggle_name', 'primary_color', 'secondary_color']],
                            left_on = 'TeamName', right_on = 'kaggle_name',
                            how = 'left')
```

### Men's teams without colors

```python
m_teams_merge[pd.isna(m_teams_merge['primary_color'])][['TeamID', 'TeamName']]
```
